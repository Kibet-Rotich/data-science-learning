introduction to seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('car_crashes')
df
```

```python
#distribution plots

#univariate distributions (one variable)

sns.displot(df['not_distracted'], kde = False, bins = 25)
```

```python
#jointplot
#plots two variables

sns.jointplot(x = 'speeding', y = 'alcohol', data = df, kind = 'reg')
```

```python
#kde plots
sns.kdeplot(df['alcohol'])
```

```python
#pair plot
sns.pairplot(df)
```

```python
tipdf = sns.load_dataset('tips')

sns.rugplot(tipdf['tip'])
```

```python
#styling

sns.set_style('white')


sns.jointplot(x='speeding', y= 'alcohol', data = df, kind = 'reg')
sns.despine(left=True,right=True, top = True, bottom= True)
```

categorical plots

```python
#bar plots
sns.barplot(x = 'sex', y = 'total_bill', data = tipdf)
```

```python
#count plots
sns.countplot(x= 'sex', data = tipdf)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tipdf, hue = 'sex')
plt.legend(loc = ())
```
