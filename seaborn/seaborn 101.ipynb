# üêß Data Visualization with Seaborn

Seaborn is a high-level data visualization library built on top of Matplotlib. It provides a simple interface for creating informative and attractive statistical graphics.

In this notebook, we‚Äôll cover:
- Univariate plots
- Bivariate plots
- Categorical plots
- Multivariate plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load a sample dataset
tips = sns.load_dataset('tips')
tips.head()
```

## 5.1 Univariate Plots
Univariate plots help us understand the distribution of a single variable.

### Histogram (sns.histplot)

```python
sns.histplot(data=tips, x='total_bill', bins=20, kde=False, color='skyblue')
plt.title('Distribution of Total Bill')
plt.show()
```

### KDE Plot (sns.kdeplot)

```python
sns.kdeplot(data=tips, x='total_bill', fill=True, color='orange')
plt.title('KDE Plot of Total Bill')
plt.show()
```

### Boxplot and Violin Plot

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
sns.boxplot(data=tips, x='day', y='total_bill', ax=axes[0])
sns.violinplot(data=tips, x='day', y='total_bill', ax=axes[1], palette='Set2')
axes[0].set_title('Boxplot of Total Bill by Day')
axes[1].set_title('Violin Plot of Total Bill by Day')
plt.show()
```

## 5.2 Bivariate Plots
Bivariate plots show relationships between two variables.

### Scatter Plot (sns.scatterplot)

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', size='size', palette='viridis')
plt.title('Total Bill vs Tip by Gender')
plt.show()
```

### Bar Plot (sns.barplot)

```python
sns.barplot(data=tips, x='day', y='tip', estimator=np.mean, ci=None, palette='pastel')
plt.title('Average Tip per Day')
plt.show()
```

### Line Plot (sns.lineplot)

```python
sns.lineplot(data=tips, x='size', y='total_bill', marker='o', color='teal')
plt.title('Line Plot of Total Bill by Size')
plt.show()
```

## 5.3 Categorical Plots
Categorical plots are useful for visualizing categorical variables.

### Count Plot (sns.countplot)

```python
sns.countplot(data=tips, x='day', palette='coolwarm')
plt.title('Count of Records per Day')
plt.show()
```

### Swarm Plot and Strip Plot

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
sns.stripplot(data=tips, x='day', y='total_bill', ax=axes[0])
sns.swarmplot(data=tips, x='day', y='total_bill', ax=axes[1])
axes[0].set_title('Strip Plot')
axes[1].set_title('Swarm Plot')
plt.show()
```

## 5.4 Multivariate Plots
Multivariate plots visualize more than two variables together.

### Pair Plot (sns.pairplot)

```python
sns.pairplot(tips, hue='sex', diag_kind='kde', palette='husl')
plt.suptitle('Pairwise Relationships in Tips Dataset', y=1.02)
plt.show()
```

### Heatmap (sns.heatmap)

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
